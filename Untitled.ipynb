**Step 2** Cleaning the dataset
1. Displaying the total number of missing values in numerical and categorical columns
2. Displaying the number of unique values in each column. If all values are different in categorical columns considered non-statistical info.
3. If the feature has too high proportion of missing values, empty rows should be removed. As replacing with mean will introduce bias to our model.

```python
# Summary of missing values in the dataset sorted as numerical columns and categorical columns and all columns
def missing_values_summary(data):
    
    missing_values = data.isna().sum()
    columns_with_missing = missing_values[missing_values > 0]
    
    numerical_cols = data.select_dtypes(include=['number']).columns
    categorical_cols = data.select_dtypes(exclude=['number']).columns
    numerical_missing = columns_with_missing[columns_with_missing.index.isin(numerical_cols)]
    categorical_missing = columns_with_missing[columns_with_missing.index.isin(categorical_cols)]
    
    missing_summary = pd.DataFrame({
        'Total Missing Values': columns_with_missing,
        'Categorical Missing Values': categorical_missing,
        'Numerical Missing Values': numerical_missing,
    }).fillna(0) 
    
    return missing_summary
    
pd.set_option('display.width', 1000)
missing_summary = missing_values_summary(data)
print ('Summary list of missing values: ', "\n", missing_summary)
```

```python
# Checking non-statistical columns 
data.nunique()
```

```python
def clean_dataset(data):
    
    # Step 1: Removing missing values more than 35%
    values_missing = data.isnull().mean() * 100
    morethan35 = values_missing[values_missing > 35].index
    data = data.dropna(subset=morethan35)
    
    # Step 2: Removing empty rows in categorical column
    columns_categorical = data.select_dtypes(include=['object']).columns
    for col in columns_categorical:
        mode = data[col].mode()[0]
        data.loc[:, col] = data[col].fillna(mode)
    
    # Step 3: Replacing missing values with mean value in numerical column
    columns_numerical = data.select_dtypes(include=['number']).columns
    for col in columns_numerical:
        mean_value = data[col].mean()
        data.loc[:, col] = data[col].fillna(mean_value)

    # Step 4: Identify and drop columns with all unique values
    unique_cols = [col for col in data.columns if data[col].nunique() == len(data)]
    data = data.drop(columns=unique_cols)
    
    return data
data_clean =  clean_dataset(data)
print("Dataframe after cleaning:", data_clean.shape)
data_clean.info()
```

# Building Logistic Regression Classification Model

```python
# Splitting the data into training(80%) and testing data (20%) and use random_state=student_id
train, test = train_test_split(data_clean, test_size=0.2, random_state=42)
print(train.shape)
print(test.shape)
```

```python
# Separating Labels and Features
x_train = train.drop(['RainTomorrow'], axis=1)
y_train = train['RainTomorrow']
x_test = test.drop(['RainTomorrow'], axis=1)
y_test = test['RainTomorrow']
# Checking data shape again after separating 
print("x_train shape: ", x_train.shape)
print("y_train shape: ", y_train.shape)
print("x_test shape: ", x_test.shape)
print("y_test shape: ", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

print("x_train_scaled shape: ", x_train_scaled.shape)
print("y_train shape: ", y_train.shape)
print("x_test_scaled shape: ", x_test_scaled.shape)
print("y_test shape: ", y_test.shape)
```

```python
# Step 2.2.1 Train a Logistic Regression Model
model = LogisticRegression(max_iter=1000)
model.fit(x_train_scaled, y_train)
```

```python
#Step 2.2.2 Calculating Accuracy score on Training and Testing data

y_train_pred = model.predict(x_train_scaled)
y_test_pred = model.predict(x_test_scaled)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Accuracy score on training set:", train_accuracy)
print("Accuracy score on testing set:", test_accuracy)

if train_accuracy > test_accuracy and (train_accuracy - test_accuracy) > 0.05:
    print("The model may be overfitting.")
else:
    print("The model does not show significant signs of overfitting.")
```
